# Part 2: Data Preprocessing

This notebook contains a script for cleaning, encoding, and combining song data obtained from the Spotify API.  It is the second notebook in my supervised learning capstone. 

### Playlist #1 - Songs I Like 

```python
# Read in liked playlist
like = pd.read_csv('liked_playlist.csv')

# Create new column 'target' column, set values to 1
like['target'] = 1

# Check it out 
like.head()
```

### Playlist #2 - Songs I Don't Like

```python
# Read in disliked playlist
dislike = pd.read_csv('disliked_playlist.csv')

# Create new 'target' column, set values to 0
dislike['target'] = 0

# Check it out
dislike.head()
```

### Merge Playlists

```python
# Specify frames 
frames = [like, dislike]

# Concatenate frames
merge = pd.concat(frames)

# Drop extra index column
merge = merge.drop('Unnamed: 0', 1)

# Confirm changes
merge.head()
```

```python
# Check data types. All columns are numeric so no changes needed.
merge.dtypes
```

### Write to CSV

```python
# Export to csv
merge.to_csv('spotify_data.csv')

# Now on to the fun part!
```
